```python
data={"User_id":[1,1,1,1,2,3,4,4,4,4,5,5,5,5,5,],
      "Product_id":["P11","P12","P13","P14","P15","P16","P17","P18","P19",
                    "PI10","PI11","PI12","PI13","PI14","PI15"],
      "Purchase":[10,20,30,50,20,40,20,30,40,50,10,20,30,40,50]}
data
```

```python
df=pd.DataFrame(data)
df
```

```python
df.groupby("User_id")["Purchase"].mean()
df
```

```python
df2=df.groupby("User_id")["Purchase"].mean().rename("User_name").reset_index()
df2
```

```python
df3=df.merge(df2)
df3
```
